```python
from selenium import webdriver as optSelenium
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pyautogui as timer
import pyautogui as buttons
from openpyxl import load_workbook
from openpyxl import Workbook
import os

browser = optSelenium.Chrome()
browser.get("https://www.ifood.com.br")

caminhoArquivo = "C:\\Users\\ITALO\\Downloads\\Planilha_Pesquisa_Ifood.xlsx"
planilha = load_workbook(filename=caminhoArquivo)
sheetPlanilha = planilha['Produtos']

# Clique input pesquisar
browser.find_element(By.XPATH, '//*[@id="__next"]/div[1]/form/div/input').click()
timer.sleep(2)
# Clique input buscar endereço
browser.find_element(By.CLASS_NAME, 'address-search-input__button').click()
timer.sleep(3)
browser.find_element(By.XPATH, '/html/body/div[4]/div/div/div/div/div/div[2]/div/div[1]/div[2]/input').send_keys("Av. Duarte da Silveira - Centro, João Pessoa, PB")

timer.sleep(2)
# Digitar endereço

resultsAddress = browser.find_elements(By.CLASS_NAME, "btn-address--full-size")
timer.sleep(3)

# Validar se endereço foi encontrado
if len(resultsAddress) > 0:
        
        timer.sleep(3)
        browser.find_element(By.XPATH, '/html/body/div[4]/div/div/div/div/div/div[2]/div/div[1]/div[3]/ul/li[1]/div/button').click()
        timer.sleep(3)
        browser.find_element(By.NAME, "addressEmptyNumber").click()
        timer.sleep(3)
        browser.find_element(By.XPATH, '/html/body/div[4]/div/div/div/div/div/div[2]/div/div[3]/div/form/button').click()
        timer.sleep(3)
        browser.find_element(By.XPATH, '/html/body/div[4]/div/div/div/div/div/div[3]/div[2]/button').click()
        timer.sleep(3)
        browser.find_element(By.XPATH, '/html/body/div[4]/div/div/div/div/div/div[3]/div[1]/div[2]/form/div[2]/div/label/input').send_keys("Sou um robô e não tenho")
        timer.sleep(3)
        browser.find_element(By.XPATH, '/html/body/div[4]/div/div/div/div/div/div[3]/div[1]/div[2]/form/div[3]/div/button[1]').click()
        timer.sleep(3)
        browser.find_element(By.XPATH, '/html/body/div[4]/div/div/div/div/div/div[3]/div[1]/div[2]/form/div[4]/button').click()
        timer.sleep(3)
        browser.find_element(By.XPATH, '//*[@id="__next"]/div[1]/header/div[2]/form/div/input').send_keys("Açaí 500ml")
        timer.sleep(3)
        buttons.press('enter')
        timer.sleep(6)
        browser.find_element(By.XPATH, '//*[@id="marmita-tab1-1"]').click()
        timer.sleep(3)

        listProducts = browser.find_elements(By.CLASS_NAME, 'merchant-list-carousel__item')

        sheetPlanilha['A1'] = "Produto"
        sheetPlanilha['B1'] = "Preço"

        linhaInicio = 2
        for info in listProducts:
              #nomeEmpresa = info.find_element(By.TAG_NAME, 'span')
              nome = info.find_element(By.CLASS_NAME, "merchant-list-carousel__item-title").text
              preco = info.find_element(By.CLASS_NAME, "card-stack-price__DEFAULT").text

              print(nome + " - R$" + preco)
              # Para cada linha, ele preenche as células na coluna A e B com o texto.  
              sheetPlanilha['A%s' % linhaInicio] = nome
              sheetPlanilha['B%s' % linhaInicio] = preco

              linhaInicio += 1

   

        planilha.save(filename=caminhoArquivo)
        os.startfile(caminhoArquivo)

else:

    print('Nada foi encontrado')


```
