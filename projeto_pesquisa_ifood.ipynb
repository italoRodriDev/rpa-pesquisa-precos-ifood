```python
from selenium import webdriver as optSelenium
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pyautogui as timer
import pyautogui as buttons
import xlsxwriter as optXlsxWriter # permite criar planilha do zero com formatação avançada
import os

browser = optSelenium.Chrome()
browser.get("https://www.ifood.com.br")

# Clique input pesquisar
browser.find_element(By.XPATH, '//*[@id="__next"]/div[1]/form/div/input').click()
timer.sleep(2)
# Clique input buscar endereço
browser.find_element(By.CLASS_NAME, 'address-search-input__button').click()
timer.sleep(3)
browser.find_element(By.XPATH, '/html/body/div[4]/div/div/div/div/div/div[2]/div/div[1]/div[2]/input').send_keys("Av. Duarte da Silveira - Centro, João Pessoa, PB")

timer.sleep(2)
# Digitar endereço

resultsAddress = browser.find_elements(By.CLASS_NAME, "btn-address--full-size")
timer.sleep(3)

# Validar se endereço foi encontrado
if len(resultsAddress) > 0:
        
        timer.sleep(3)
        browser.find_element(By.XPATH, '/html/body/div[4]/div/div/div/div/div/div[2]/div/div[1]/div[3]/ul/li[1]/div/button').click()
        timer.sleep(3)
        browser.find_element(By.NAME, "addressEmptyNumber").click()
        timer.sleep(3)
        browser.find_element(By.XPATH, '/html/body/div[4]/div/div/div/div/div/div[2]/div/div[3]/div/form/button').click()
        timer.sleep(3)
        browser.find_element(By.XPATH, '/html/body/div[4]/div/div/div/div/div/div[3]/div[2]/button').click()
        timer.sleep(3)
        browser.find_element(By.XPATH, '/html/body/div[4]/div/div/div/div/div/div[3]/div[1]/div[2]/form/div[2]/div/label/input').send_keys("Sou um robô e não tenho")
        timer.sleep(3)
        browser.find_element(By.XPATH, '/html/body/div[4]/div/div/div/div/div/div[3]/div[1]/div[2]/form/div[3]/div/button[1]').click()
        timer.sleep(3)
        browser.find_element(By.XPATH, '/html/body/div[4]/div/div/div/div/div/div[3]/div[1]/div[2]/form/div[4]/button').click()
        timer.sleep(3)
        browser.find_element(By.XPATH, '//*[@id="__next"]/div[1]/header/div[2]/form/div/input').send_keys("Açaí 500ml")
        timer.sleep(3)
        buttons.press('enter')
        timer.sleep(6)
        browser.find_element(By.XPATH, '//*[@id="marmita-tab1-1"]').click()
        timer.sleep(3)

        listProducts = browser.find_elements(By.CLASS_NAME, 'merchant-list-carousel__item')
        
        caminhoArquivo = "C:\\Users\\ITALO\\Downloads\\Planilha_Pesquisa_Ifood.xlsx"
        planilha = optXlsxWriter.Workbook(caminhoArquivo)
        sheetPlanilha = planilha.add_worksheet("Dados")
        
        # Alterar estilo do cabeçalho
        styleColum = planilha.add_format({'align': 'left',
                                 'font_color': 'white',
                                 'bold': True,
                                 'bg_color': '#8000ff'})
        
        # Alterar estilo da linha
        styleLine = planilha.add_format({'align': 'left',
                                 'font_color': 'white',
                                 'bold': True,
                                 'bg_color': '#00ff80'})
        
        sheetPlanilha.write("A1", "Produto", styleColum)
        sheetPlanilha.write("B1", "Preço", styleColum)
        sheetPlanilha.write("C1", "Loja", styleColum)
      
        # Localizar o elemento <h1> com a classe 'merchant-list-carousel__merchant-title'
        nome_empresa_element = browser.find_elements(By.CLASS_NAME, 'merchant-list-carousel__merchant-title')
        # Dentro deste h1, encontrar o <span> e extrair o texto
        enterprises = []
        for i in nome_empresa_element:
            span_element = i.find_element(By.TAG_NAME, 'span')
            nome_empresa = span_element.text
            if nome_empresa in enterprises:
                print('já existe')
            else:

                enterprises.append(nome_empresa)
        
        linhaInicio = 2
        for info in listProducts:
              nome = info.find_element(By.CLASS_NAME, "merchant-list-carousel__item-title").text
              preco = info.find_element(By.CLASS_NAME, "card-stack-price__DEFAULT").text
              if preco is not '':
                preco1 = preco.split("R$")
                preco2 = preco1[1].replace(".", "").replace(",", ".")
                precoFormatado = float(preco2)


                # Para cada linha, ele preenche as células na coluna A e B com o texto.  
                sheetPlanilha.write('A%s' % linhaInicio, nome, styleLine)
                sheetPlanilha.write('B%s' % linhaInicio, precoFormatado, styleLine)
                sheetPlanilha.write('C%s' % linhaInicio, "x", styleLine)
                linhaInicio += 1

        graficoProdutos = planilha.add_chart({'type': 'column'})

        graficoProdutos.add_series({
            'name': '=Dados!A$1',
            'categories': '=Dados!$A$2:$A$10',
            'values': '=Dados!$B$2:$B$10'
        })

        graficoProdutos.set_title({'name': 'Quanto seus concorrentes estão cobrando'})
        graficoProdutos.set_x_axis({'name': 'Produtos'})
        graficoProdutos.set_y_axis({'name': 'Preços'})

        graficoProdutos.set_style(11)

        sheetPlanilha.insert_chart('D2', graficoProdutos, {'x_offset': 25, 'y_offset': 10})

        planilha.close()

        os.startfile(caminhoArquivo)

else:

    print('Nada foi encontrado')


```
